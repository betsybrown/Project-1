```python
wine_data_sample = wine_data_sample.reset_index(drop=True)
#country1=wine_data_sample["country"][ticker]
#print(country1)
```

```python
places = list(wine_data_sample["winery"])
listofcountries = list(wine_data_sample["country"])

lat = []
long = []
ticker=0
country1=wine_data_sample["country"][ticker]
for place in places:

    #base_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"
    url = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={place}%20winery%20{country1}&inputtype=textquery&fields=formatted_address,name,geometry&key={api_key}"
    ticker+=1
    response = requests.get(url).json()
    if response["status"] == "OK":
            #print(json.dumps(response, indent=4, sort_keys=True))
            print(f"Processing Record {ticker} of {n} |")
            print(response["candidates"][0]['geometry']['location']['lat'])
            lat.append(response["candidates"][0]['geometry']['location']['lat'])
            long.append(response["candidates"][0]['geometry']['location']['lng'])
            #print(f"Processing Record {ticker} of {n} | lat:{city} long:{}")
            #time.sleep(.000005)
    else:
            lat.append("NaN")
            long.append("NaN")
            print("Skipping to next place.")
time.sleep(.00005)
    
print("--------------------------")
print("Data Retrieval is Complete")
print("--------------------------")
```

```python
# Dependencies
from citipy import citipy
```

```python
result = zip(lat,long)
coordinates = list(result)
type(coordinates[0][0])
```

```python
cities = []
for coordinate_pair in coordinates:
    lat1, long1 = coordinate_pair
    if type(coordinate_pair[0]) is float and type(coordinate_pair[1]) is float:
        city1 = citipy.nearest_city(lat1, long1)
        cities.append(city1.city_name)
    else:
        cities.append("No Info")
```

```python
# Save config information.


url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
api_key = "9d685dea3c99b3a25afb99d8a3ffe4b6"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
# set up lists to hold reponse info
citydict = []

cityname = []
cloud = []
country = []
date = []
humid = []
lat = []
lng = []
maxtemp = []
windspeed = []

counter = 1
setnum = 1
print("Beginning Data Retrieval")
print("------------------------")

# Loop through the list of cities and perform a request for data on each
for city in cities:
    
    print(f"Processing Record {counter} of Set {setnum} | {city}")
    
    counter = counter + 1
    
    response = requests.get(query_url + city)
    data = response.json()
    
    if(response.status_code == 200):
        
        cityname.append(data["name"])
        cloud.append(data["clouds"]["all"])
        country.append(data["sys"]["country"])
        date.append(data["dt"])
        humid.append(data["main"]["humidity"])
        lat.append(data["coord"]["lat"])
        lng.append(data["coord"]["lon"])
        maxtemp.append(data["main"]["temp_max"])
        windspeed.append(data["wind"]["speed"])
        
        citydict = {"City": cityname,
                    "Cloudiness": cloud,
                    "Country": country,
                    "Date": date,
                    "Humidity": humid,
                    "Lat": lat,
                    "Lng": lng,
                    "Max Temp": maxtemp,
                    "Wind Speed": windspeed}

    else:
        cityname.append("NaN")
        cloud.append("NaN")
        country.append("NaN")
        date.append("NaN")
        humid.append("NaN")
        lat.append("NaN")
        lng.append("NaN")
        maxtemp.append("NaN")
        windspeed.append("NaN")
        
        print(f"City:{city} was not found. | Retrieving data for next entry.")
        setnum = setnum + 1
    time.sleep(.0005)
   

print("--------------------------")
print("Data Retrieval is Complete")
print("--------------------------")
```
